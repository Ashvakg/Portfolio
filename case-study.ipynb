Let's use some SQL Queries to check if we can actually merge the dataframes.


```r
#Checking for Daily_activity and daily_calories after seeing the head function i came to a conclusion that both the datasets have certain colums of same name and same data.So we join them together but before that let's make a quick check if we're correct.
#Checking the column name of daily_calories -> Activity Day to Activity Date.
colnames(daily_activity)[2] <- "ActivityDay"

head(daily_activity)


```

**We have changed the column name to Activity day so we can merge our dataframe**

```r

daily_activity2 <- daily_activity %>%
  select(Id, ActivityDay, Calories)

head(daily_activity2)
```

```r
#Checking with SQL(daily_activity and daily_calories)
library(sqldf) # for sql queries
sql_check1 <- sqldf('SELECT * FROM daily_activity2 INTERSECT SELECT * FROM daily_calories')
head(sql_check1)
nrow(sql_check1)
```

```r
plot(density(calories.res))
```
